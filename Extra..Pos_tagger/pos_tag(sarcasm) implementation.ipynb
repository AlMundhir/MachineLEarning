```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tree import *
from nltk.tokenize import RegexpTokenizer
import pandas as pd
```

```python
# tweets_corpus = pd.read_csv('/Users/adhraa/Desktop/uni/ADS/Coosto/dataset-4412-cleaned.csv',sep="\n")
df = pd.read_csv("/Users/muntheerkiyumi/Desktop/Minor/Project/Tweets-all.csv", delimiter='|', encoding="utf-8", header=None, names=['TweetID','Tweet','Status','positive','neurta','negative','capitals','pos_tag'])

df
```

```python
# np="NP: {<DT>?<RB>?<JJ>*<NN>}"
# adjp="ADJP: {<RB>*<RB>?<RB>}"
# vp="VP: {<RB><MD><CC>}"
#Created after many tries and implementation.. we came to conculde that we need this sequence..
int2="INT: {<UH>?<.>(<RB><JJ>|<JJ><N>|<RB><V>|<JJ>|<RB>)}"
sarcastic=[]  
```

```python
# sentence=["Oh! niet te vreselijk","Wow, nu dat ik weet over Photoshop, is mijn jarenlange eetstoornis genezen!",
#           "eindelijk is het vrijdag"]
for tweet in df['Tweet']:
	tweet_tag=nltk.pos_tag(word_tokenize(tweet))
	cp=nltk.RegexpParser(int2)
	results=cp.parse(tweet_tag)
# 	print(results)
	for x in results[:3]:
# 		print(x)        
		if isinstance(x,nltk.tree.Tree):
			print(x)
			if x.label()=='INT':
# 				sarcastic.append(x)
				print("sarcastic",x,tweet)
# print(sarcastic)
```
